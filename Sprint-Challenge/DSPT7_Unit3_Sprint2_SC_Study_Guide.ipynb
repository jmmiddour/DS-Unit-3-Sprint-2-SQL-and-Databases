## SQL

**Concepts:**

1. What is SQL?
	- Structured Query Language (SQL)
	- Is the standard language for interating with (accessing and manipulate) databases using queries.
	- Became a standard of the American National Standards Institue (ANSI) in 1986 and the International Organization for Standardization (ISO) in 1987.
	- There are different versions of SQL such as;
		- MySQL 
			- The most popular in Stack Overflow. 
			- Similar in syntex to SQLite and PostgreSQL.
		- SQLite 
			- Similar in syntex to MySQL and PostgreSQL.
		- PostgreSQL 
			- Best starter version of SQL.
			- Similar in syntex to MySQL and SQLite.
			- The syntex most closely conforms to Standard SQL
		- SQL Server
			- An important database management system in many industries.
		- T-SQL 
			- The name of Microsoft's dialect of SQL. Has the greatest contrast in SQL syntex and a wide variety of functions not available in other platforms.

2. What is a RDBMS?
	- Rational DataBase Management System (RDBMS)
	- A Rational DataBase is a type of database that uses structure that allows you to identify and access data in relation to another piece of data in the database. 
	- Most often is a relational database that is organized into tables
	- Management System is a program that allows you to create, update, and administer a rational database.
	- Most RDBMS use SQL language to access the database.
3. What is an ETL pipeline?
	- Extract Transform Load (ELT)
	- A set of processes extracting data from an imput source, then transforming the data and then finally loading the data into an output destination, such as a database, to use for reporting, analysis, and data synchronization.
4. What is a schema?
	- A map of the database. 
	- Explains attributes or columns in each table within the database.
	- Provides the primary and foriegn keys for each table to show their ralationship to each other.
	- Sets the requirements of the type of data that can be stored in each column.
5. What does each letter in ACID stand for? Give an explanation for each and why they matter?
	- **A** = Atomicity
		- Guarantees that each transaction goes through completely or the transaction fails as a whole.
		- Each "transaction" is counted as a single unit of a whole.
	- **C** = Consistancy
		- If all values are not formatted correctly it will not add the record at all.
		- All attribrutes must be met.
		- Upon completion of a transaction the database is structurally sound.
	- **I** = Isolation
		- Ensures that concurrent execution leaves the database in the same state that would have been obtained in they were executed seperately.
		- Access to the data is moderated by the database so that transactions appear to run sequentially.
	- **D** = Durability (Fault Tolerance)
		- Guarantees that once a record has been commited, it will remain commited even in the case of a system failure.
	- Best for using when dealing with data that has to be consistant at all times. No room for error, such as banking institutes.
6. Explain each of the table relationships and give an example for each
	- One-to-One = There will only be one record in the first table (primary) that relates to only one record in the second table (foreign).
	- One-to-Many = There will only be one record in the  first table (primary) that relates to zero, one, or multiple records in the second table (foreign).
	- Many-to-Many = There can be zero, one, or multiple records in the first table (primary) that relates to zero, one, or multiple records in the second table (foreign).

### Questions of Understanding

1. What is a document store?
  - Document-oriented database = A computer program designed for storing, retrieving and managing document-oriented information.

2. What is a `key:value` pair?
  - When you have a value that you assign to a variable (key)
  - Your "key" is where the value will be stored.
  - Example: In a DataFrame the column would be your "key" and the value is what is in that column.
  - Useful for mapping to key variables.
- What data type in Python uses `key:value` pairs?
  - Dictionaries

3. Give an example of when it would be best to use a SQL Database and when it would be best to use a NoSQL Database
  - **SQL** (Relational or RDBMS (_SQLite_ or _PostgreSQL_) Databases):
    - Smaller databases (application) that does not involve loads of querying and needs a structured schema to organize the data more clearly for readability.
    - _Examples_: 
      - A company that has less then 1000 employees and they need to keep track of what each employee does and their information.
      - An application that can allow for redundancy and involves less joins.
  - **NoSQL** ("Not Only SQL" or "Not SQL" (Non-relational or Document-Oriented (_MongoDB_) Databases)):
    - Larger databases (application) that involves loads of querying and do not require the need for a structured schema to organize the data.
    - _Examples_: 
      - Facebook, Google, or Instagram user database. 
      - Applications that involve atomic transactions.

4. What are some of the trade-offs between SQL and NoSQL?  
  - **SQL** ((Relational or RDBMS (_SQLite_ or _PostgreSQL_) Databases):
    - **(+)** More intuitive in terms of connecting, standardized ways to interact.
    - **(+)** Readability of SQL and documentation for wrapper libraries.
    - **(+)** Consistency guarantees in terms of columns/data.
    - **(+)** Simple to use since everything exsist in a single server meaning there is no need to manage multiple instances.
    - **(+/-)** Requires the use of a predefined Schema to keep the data organized clearly.
    - **(-)** Vulnerable to SQL injections (security vulnerabilities)
    - **(-)** Requires Vertical scaling/ scale up (requires more physical or virtual resources because everything is stored on one machine, so have to have a machine with enough resources to handle the workload).
    - **(-)** Harder to scale beyond ~10-100 TB.
    - **(-)** Difficult to preform multiple queries simultaneously.
    - **(-)** Chances of downtime are high, when the server exceeds maximum load.
  - **NoSQL** ("Not Only SQL" or "Not SQL" (Non-relational or Document-Oriented (_MongoDB_) Databases)):
    - **(+)** Easier for use cases like scraping web data.
    - **(+)** Great for less structured data.
    - **(+)** Horizontal scaling (does not require as many resources as SQL because it is scaled by adding more machines to help with the workload.)
    - **(+)** Easier to scale beyond ~10 TB or more.
    - **(+)** It is cheaper because less resources are needed.
    - **(+)** Lesser load on machines which equals better preformance.
    - **(+)** Chances of less downtime.
    - **(+)** Resilience and Fault Tolerance.
    - **(+/-)** Does not follow the concept of foreign key or normalization.
    - **(+/-)** Uses a dynamic Schema for unstructured data.
    - **(-)** Can potentially lead to very messy data by not requiring specific schema.
    - **(-)** Less suitable for RDBMS as it relies on strict Consistency and Atomicity rules.
    - **(-)** Making joins is more diffuicult because it may involve cross-server communication.
    - **(-)** Can't easily categorize each feature to each server. Sometimes images may take up more space than a single server can handle, therefore, splitting it between multiple machines.
  - **Considerations** when choosing best one to use:
    - **Skill set** of the team doing the work (SQL pro or Newbie)
    - **Possible growth** in terms of:
      - _Volume_: Total storage size needed.
      - _Variety_: Types of documents or data.
      - _Velocity_: How quickly it is coming in.
    - **Uptime requirements** (SLA - Service Level Agreement)
    - **Error handling**

5. What does each letter in BASE stand for? Give an explanation for each and why they matter?
    - **B** = Basic:
      - Combines with A (Availability)...
    - **A** = (Basic) Availability:
      - The database appears to work most of the time.
      - This gives an advantage to NoSQL databases because the data is spread across many storage systems with a high degree of replication. 
      - If there is a failure in a segment of the data, this does not necessarily result in a complete database outage.
    - **S** = Soft-state:
      - Stores that don't have to be write-consistent.
      - Different replicas don't have to be mutually consistent all the time.
      - This leaves the data consistency in the developer's hands and is not handled by the database.
    - **E** = Eventual consistency:
      - Stores exhibit consistency at some later point (e.g. lazily at read time), no guarantess are made as to when this will occur.
    - **BASE** properties are much looser than ACID guarantees, but there isn't a direct one-for-one mapping between the two consistency models.
    - **BASE** consistency is important for scaling because it values availability, but doesn't offer guaranteed consistency of replicated data at write time.
    - Data will be consistent in the future, either at read time or it will always be consistent, but only for certain processed past snapshots.
    - Developers need to be more knowledgeable and rigorous about consistent data.
    - The **BASE** model is a flexible alternative to the **ACID** model for databases that don't require strict adherence to a rational model.
    - **BASE** is better suited for "fuzzy" subjects like sentiment analysis and data that is not well-structured.


## Data Normalization:

A type of process wherein the data within a _**relational database**_ is reorganized in such a way that users can properly utilize that database for further queries and analysis. This is to reduce data redundancy and improve data integrity.

- There are 2 primary advantages of having a highly normalized data schema:
  1. **Increased Consistency**: Information is stored in one placec only, reducing the possibility of inconsistent data.
  2. **Easier Object-to-Data Mapping**: Highly-normalized data schemas in general are closer conceptually to object-oriented schemas because the object-oriented goals of promoting high cohesion and loose coupling between classes results in similar solutions from a data point of view.

- The primary disadvantage is slower reporting performance. 

- The main goal is... 
  - Whenever there's lots of duplication within a table, try to seperate that table and identify the relationship between some primary key in the main table and a foreign key in the other table.
  - Working within a smaller table and changing a value once and then joining tables gives you a more unified view of the data more regularly.
